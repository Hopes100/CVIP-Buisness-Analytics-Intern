```python
df = df.drop_duplicates()
df
```

```python
df = df.drop(columns = 'Culture & Values')
df
```

```python
df = df.drop(columns = 'Business Outlook')
df
```

```python
df["cons"] = df["cons"].str.replace('#NAME?', '')
df
```
